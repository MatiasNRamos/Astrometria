En este problema se ve el tema del cambio de variable aleatoria. ¿Cómo es la distribución de probabilidad de una variable Y(x) si conozco la de x?. El teorema me dice cómo calcularla.

En el código genero números que sigan la función Y = cos(x) y grafico su histograma para compararlo con la distribución de probabilidad teórica y ver si coinciden

```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt
# Estilo de gráficos
plt.style.use('dark_background')
```

```python
plt.style.use('dark_background')
```

```python
# Genero puntos equiprobables de distancia cenital
N = 20000
Nums = np.pi*np.random.random(N)
Cos = np.cos(Nums)
# Materializo la distribución de probabilidad:         
Ys = np.linspace(-1,1,1000)
fy = 1 * abs( np.tan(Ys)/np.cos(Ys) ) # Distribución de probabilidad de Y=cos(x)

# Veo los gráficos
fig, ax = plt.subplots(1, 1, figsize = (12,6))

ax.hist(Cos, color='red', bins=50, density=True, label='Histograma generado')
ax.plot(Ys, fy, color='yellow', label='Función de probabilidad teórica')

ax.set_title('Distribucion Y=cos(x)', fontsize=20)
ax.set_xlabel('N° (valor de Y)', fontsize=20)
ax.set_ylabel('Frecuencia', fontsize=20);
ax.legend(fontsize=20);
```
