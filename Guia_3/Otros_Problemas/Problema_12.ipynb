```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm
# Estilo de gráficos
plt.style.use('dark_background');
```

```python
plt.style.use('dark_background')
# Creo puntos de la gaussiana
N = 10000
# La desviación estándar asociuada depende de la muestra tomada de 50 puntos
G = norm.rvs(loc=1550, scale=30/np.sqrt(50), size=N)
# Para saber el intervalo de confianza de 95% uso los cuantiles 0,025 y 0,975
q1 = norm.ppf(q=0.025, loc=1550, scale=30/np.sqrt(50))
q2 = norm.ppf(q=0.975, loc=1550, scale=30/np.sqrt(50))
```

```python
# Grafico del intervalo de confianza del 95%
fig, ax = plt.subplots(1, 1, figsize = (12,6))

ax.hist(G, color='green', density=True)
# Cuantiles
ax.axvline(x=q1, ymin=0, ymax=1, color='orange', ls='--', lw=5, label='I.C. del 95%')
ax.axvline(x=q2, ymin=0, ymax=1, color='orange', ls='--', lw=5)

ax.set_title('Gaussiana', fontsize=20)
ax.set_ylabel('Frecuencia relativa', fontsize=20)
ax.set_xlabel('x', fontsize=20)
ax.legend(fontsize=15);
```

```python
print('Intervalo de confianza del 95%:', '(', round(q1,1), ', ', round(q2,1), ') horas')
```
