En este programa se grafican los datos de blanton para la función luminosidad

El código sólo importa los datos del .csv y los grafica

```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt
import pandas
# Estilo de gráficos
plt.style.use('dark_background')
```

```python
plt.style.use('dark_background')
Datos = pandas.read_csv('blanton.csv', sep=',')
# Pongo los datos en dos variables
Mags = Datos['M']
Lum = Datos['f']
Barra_sup = np.loadtxt('Barra_sup.txt')
Barra_inf = np.loadtxt('Barra_inf.txt')

ERR = []
ij = 0
while ij<len(Barra_sup):
    ERR.append( (1/2.) * (Barra_sup[ij] - Barra_inf[ij]) )
    ij = ij + 1
```

```python
fig, ax = plt.subplots(1, 1, figsize = (12,6), sharex=True)

ax.errorbar(Mags, Lum, yerr=ERR, color='cyan', fmt='.', label='Datos')

ax.set_xlabel('Magnitud', fontsize=20)
ax.set_ylabel('Luminosidad', fontsize=20)
ax.set_yscale('log')
ax.legend(fontsize=15, loc=2);
```

A las barras de error se las obtuvo usando un programa para sacar datos de las figuras. Las barras grandes al final del plot son algunos artefactos de la representación logarítmica, las barras tienen el valor correcto. (Ver problema 3 para más detalles de ésto último)
